```python
import re
```

```python
def makeitstring(Text):
    Texttoreturn=str(Text)
    Texttoreturn=re.sub('[^a-zA-Z]',' ',Texttoreturn)
    Texttoreturn=Texttoreturn.lower()
    Texttoreturn=Texttoreturn.split()
    return Texttoreturn
```

```python
song['track_name']=song['track_name'].apply(makeitstring)
```

```python
song['track_album_name']=song['track_album_name'].apply(makeitstring)
```

```python
song.head()
```

```python
def collapse(L):
    L=str(L)
    L=L.replace(" ","")
    return L
```

```python
song['track_artist']=song['track_artist'].apply(collapse)
song['track_artist']=song['track_artist'].apply(lambda x:x.split())
```

```python
song.head()
```

```python
song['tags']=song['track_name']+song['track_artist']+song['track_album_name']+song['playlist_name']+song['playlist_genre']+song['playlist_subgenre']
```

```python
song_words=song[['track_id','track_album_id','playlist_id','tags']]
```

```python
song_words.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
import nltk
```

```python
porter=PorterStemmer()
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def portit(T):
    stop_words = set(stopwords.words('english'))

    # Apply Porter stemming and filter out stopwords
    stemmed_tokens = [porter.stem(token) for token in T if token.lower() not in stop_words]

    # Join the stemmed tokens into a string
    ss = ' '.join(stemmed_tokens)

    return ss
    
```

```python
song_words['tags']=song_words['tags'].apply(portit)
```

```python
song_words.head()
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(song_words['tags'])
```

```python
vector=vector.toarray()
```

```python
vector.shape
```

```python
print(cv.get_feature_names_out())
```

```python
word_similarity=cosine_similarity(vector)
```
